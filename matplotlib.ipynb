```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.arange(0,10)
y =x*x
plt.scatter(x,y,c='r')
plt.show()
```

```python
plt.plot(x,y,'r')
plt.show()
```

```python
plt.plot(x,y,'-->')
plt.title('2d graph')        
plt.show()
```

```python
plt.subplot(3,3,1)
plt.plot(x,y,'r')
plt.show()
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.show()
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.show()
```

```python
x=np.arange(0,4 *np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.plot(x,y_sin,'r--')
plt.plot(x,y_cos,'b--')
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
plt.title('cosine')
plt.show()
```

```python
#bar plot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,color = 'b')
plt.bar(x2,y2,color = 'g')
plt.title('bar graph')
plt.show()
```

```python
a=np.array([25,87,5,43,56,73,55,54,2,3,51])
plt.hist(a)
plt.title("histogram")
plt.show()
```

```python
#pie chart
labels= 'python','c++','ruby','java'
sizes = [215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0)

plt.pie(sizes,explode=explode,labels = labels,colors=colors)
plt.title('pie chart')
plt.show()
```

```python
#seaborn tutorial
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
df.head()
df['sex_num'] = df['sex'].map({'Male': 0, 'Female': 1})
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['sex_num'] = le.fit_transform(df['sex'])
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
plt.show()
```
